```python
import cv2
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Загрузка ищображений
image = cv2.imread('img.jpg')
pattern_image = cv2.imread('pattern.jpg')

# Поиск шаблона
result = cv2.matchTemplate(image, pattern_image, cv2.TM_CCOEFF_NORMED)

# Установка порога для определения совпадений
threshold = 0.8
locations = np.where(result >= threshold)

# Создание копии изображения
result = np.copy(image)

# Выделяем объекты
for loc in zip(*locations[::-1]):
    cv2.rectangle(result, loc, (loc[0] + pattern_image.shape[1], loc[1] + pattern_image.shape[0]), (0, 255, 0), 2)

plt.figure(figsize=(15,20))
plt.subplot(1, 3, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 2)
plt.imshow(cv.cvtColor(pattern_image, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 3)
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```
