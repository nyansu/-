```python
import cv2
import numpy as np

def negate_image(image):
    # Преобразуем изображение в массив пикселей
    img_array = cv2.imread(image, cv2.IMREAD_COLOR)
    # Преобразуем каждый канал в негатив
    for channel in range(3):
        img_array[..., channel] = 255 - img_array[..., channel]
    # Возвращаем обработанное изображение
    return img_array

def main():
    # Путь к исходному изображению
    image_path = 'image.jpg'
    # Загружаем изображение
    image = negate_image(image_path)
    # Сохраняем обработанное изображение
    cv2.imwrite('negate_image.jpg', image)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()

main()
```
